## let's remember the distribution:

```python
import matplotlib.pyplot as plt

# Count the number of occurrences 
smoker_counts = df['Smoking'].value_counts()

# Create a bar plot
plt.figure(figsize=(6, 4))
smoker_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Smoker Status')
plt.xlabel('Smoker Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a crosstab between 'Smoking' and 'Heart Disease'
cross_tab = pd.crosstab(df['Smoking'], df['Heart Disease'])

# Plot the grouped bar chart
cross_tab.plot(kind='bar', figsize=(8, 6), color=['skyblue', 'salmon'])
plt.title('Smoking Status vs Heart Disease')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Heart Disease', labels=['No Heart Disease', 'Heart Disease'])
plt.show()
```

### And in percentages.

```python
import pandas as pd

# Create a crosstab to count occurrences of 'Smoking' and 'Heart Disease'
cross_tab = pd.crosstab(df['Smoking'], df['Heart Disease'])

# Calculate the percentage of heart disease within each smoking category
percentage_heart_disease = cross_tab.apply(lambda row: row[1] / row.sum() * 100, axis=1)

# Print the percentages
print("Percentage of Heart Disease in each Smoking category:")
print(percentage_heart_disease)
```
